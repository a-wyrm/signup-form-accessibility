```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
main_data = pd.read_csv("./violations_checker.csv")
main_data = main_data.sort_values('Count', ascending=False)
```

```python

colors_map = {
    "critical": "#DC267F",
    "serious": "#FE6100",
    "moderate": "#FFB000",
    "minor": "#785EF0" }


#color=main_data['Severity'].map(colors_map)

plt.figure(figsize=(8,8))
plt.barh(main_data["Violation ID"], main_data["Count"], color=main_data['Severity'].map(colors_map))


patches = [ plt.Rectangle((0, 0), 1, 1, color=c) for c in ["#DC267F", "#FE6100", "#FFB000", "#785EF0"]]
plt.legend(patches, ["Critical", "Severe", "Moderate", "Minor"], title="", fontsize=9)

plt.xlabel('Number of Violations')
plt.ylabel('Name of Violation')

plt.savefig('violations.pdf', format="pdf", bbox_inches="tight" )
plt.show()
```
